### **6.3 Removing Symbols (Special Characters) From the Data**

- The feature **Size** in Play Store Data has Special Characters.

  - **'M' Represent MB(MegaByte)**: 1 MB = 1024 KB

- The feature **Price** in Play Store Data **Contains Dollar Sign ($).**

- The feature **Download_Nums** contains **'+' and ',' Symbols.**

```python
# Fixing Size Feature of Play Store Data

# Checking whether there is any other character then 'M' in the data
df_play[~df_play['Size'].str.contains('M')]['Size'].head(10)
```

**Observation:**

- There is **'k'** in the data that **Represent 'KB' (Kilobytes).**

```python
# Check whether there is any other character then 'M' and 'k' in the data

df_play[~df_play['Size'].str.contains('M|k')]['Size']
```

**Observation:**

- Every **Value in Size contains either 'k' or 'M'.**

- Therefore, we can:

  - **Remove 'k'** and keep the number as it is.
  - **Remove 'M' and Multiply the number by 1024** to convert it to KB

```python
# Fixing Size Feature of Play Store Data

def Correct_Size(size):

  if 'M' in size:
    size = size.replace('M', '')
    return float(size) * 1024

  else:
    size = size.replace('k', '')
    return float(size)

df_play['Size'] = df_play['Size'].apply(Correct_Size)

df_play[['Size']].head(10)
```

```python
# Checking the dtype of Size after Operation
df_play['Size'].dtype
```

**Observation:**

- Data in the **Size feature has been fixed.**

```python
# Fixing Price Feature of Play Store Data

df_play['Price'] = df_play['Price'].apply(lambda x : float(x.replace('$', '')))

df_play[['Price']].head(10)
```

```python
# Checking the Dtype of Price Column

df_play['Size'].dtype
```

**Observation:**

- Data in the **Price feature has been fixed.**

```python
# Fixing Download_Nums Feature

def Correct_Download_Num(Num):

  Num = Num.replace('+', '')
  Num = Num.replace(',', '')

  return int(Num)

df_play['Download_Num'] = df_play['Download_Num'].apply(Correct_Download_Num)

df_play[['Download_Num']].head(10)
```

```python
# Checking the Dtype of Download_Num Column

df_play['Download_Num'].dtype
```

**Observation:**

- Data in the **Download_Num feature has been fixed.**

### **6.4 Handling Missing Data**

- There was **Missing Data in Play Store Dataset.**

```python
df_play.columns.values
#df_play.isnull().sum().values
```

```python
# Finding Missing Data in Play Store Dataset

null_frame = pd.DataFrame(index = df_play.columns.values)
null_frame['Null Frequency'] = df_play.isnull().sum().values

percent = df_play.isnull().sum().values/df_play.shape[0]
null_frame['Missing %age'] = np.round(percent, decimals = 4) * 100

null_frame
```

```python
# Creating canvas

canv, axes = plt.subplots(1,2)
canv.set_size_inches(14,6)
canv.tight_layout(pad=5.0)
canv.suptitle('Distribution of Segment (Play Store v.s. App Store)', size=15)

# Google Play Store
plt.sca(axes[0])
sns.countplot(data=df_play, x='Segment')
for p in axes[0].patches:
  x=p.get_bbox().get_points()[:,0]
  y=p.get_bbox().get_points()[1,1]
  axes[0].annotate('{} ({:.1f}%)'.format(y, 100.*y/df_play.shape[0]), (x.mean(), y),
            ha='center', va='bottom', size=12)
sns.despine()
plt.xlabel('Segment in Google Play Store', size=12)

# IOS APP Store
plt.sca(axes[1]) # Selecting Second Subplot
sns.countplot(data=df_ios, x='Segment')
for p in axes[1].patches:
  x=p.get_bbox().get_points()[:,0]
  y=p.get_bbox().get_points()[1,1]
  axes[1].annotate('{} ({:.1f}%)'.format(y, 100.*y/df_ios.shape[0]), (x.mean(), y),
            ha='center', va='bottom', size=12)
sns.despine()
plt.xlabel('Segment in IOS App Store', size=12);
```

### **Question 9:** What is the relation between **Size of an App** and the **Number of Download** it recieves.

- Downloading an app from a store takes both **time and bandwith.**

- So we can assume that apps with **small size will have a higher number of downloads.**

- But we have to **verify this claim** to **decide size bracket for our app.**

```python
# Binning our Continous Feature "Size"

df_play['Size_Bin'] = pd.cut(df_play['Size'], bins=5, right=False)
df_play.head()
```

```python
# Creating canvas
fig = plt.figure(figsize=(14, 8))

# Plotting
Size_Downloads = df_play.groupby('Size_Bin').mean()['Download_Num'].reset_index()
sns.barplot(data=Size_Downloads, x= 'Size_Bin', y= 'Download_Num', palette='Set2')
sm = 0
for p in plt.gca().patches:
  x=p.get_bbox().get_points()[:,0]
  y=p.get_bbox().get_points()[1,1]
  plt.gca().annotate(int(y), (x.mean(), y),
            ha='center', va='bottom', size=12)
  sm = sm + int(y)

# Aesthetics
plt.title('Size of an app vs Number of Downloads', fontsize=16, pad=10)
plt.ylabel('Average Number of Downloads', size=14, labelpad=7)
plt.xlabel('Binned Sizes of the Apps', size=14, labelpad=7)
plt.tick_params(labelsize=12)
sns.despine()
```

**Observation**

- Our **Original Hypothesis** was Incorrect.

- **We can not say** that the **smaller the apps is the higher the downloads will be.**

- The **Size Bracket that we recommened** for the new app is **( [61443 - 81921) ) KBs.** or **( [60 - 80) )** MB's
